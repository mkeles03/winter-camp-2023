```python
X = []
Y = []

for x,y in points:
    X.append(x)
    Y.append(y)
    
X=np.array(X)
Y=np.array(Y)

    
    
```

```python
X1 = np.c_[np.ones(len(X)), X]
```

```python
Xsquared = X**2
X2 = np.c_[X1,Xsquared]
```

```python
def loss(w):
    return sum((np.dot(w,X2.T) - Y)**2) / len(X2)
    
def stochastic_dF(w, i):
    return (w.dot(X2[i].T)- Y[i])*X2[i]
    
def stochastic_gradient_descent(loss, stochastic_dF, d, n):
    learning_rate = 0.0005
    w = np.random.rand(X2.shape[1])
    for a in range(1000):
        for b in range(len(X1)):
            print(w, loss(w))
            i = np.random.randint(0,100000)
            gradient = stochastic_dF(w,i)
            w = w - learning_rate * gradient
        
stochastic_gradient_descent(loss, stochastic_dF, 6, len(points))
```
